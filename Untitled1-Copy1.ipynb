```python
import pandas as pd
import numpy as np

testCSV = './test_market_returns_brl.csv'
stockRows = 1243
stockCols = 63

dfStock = pd.read_csv(testCSV, nrows=stockRows)
dfStock.head()
```

```python
# cleaning the data
dfStock = dfStock.dropna(axis=0)
stockData = dfStock.iloc[0:, 2:]
# marketData = dfStock.iloc[0:, [1]]
# stockData = stockData.dropna(axis=0)
# stockData.head()

# update das cols e rows depois da limpeza
[stockRows, stockCols] = stockData.shape
stockData
# print(stockRows, stockCols)
# [marketRows, marketCols]=marketData.shape
# print(marketRows, marketCols)
```

```python
# retorna o return de uma stock específica
# def stockReturns(stockPrice, numRows, numColumns):    
#     stockReturns = np.zeros([numRows-1, numColumns])
#     for j in range(numColumns):
#         for i in range(numRows-1):
#             stockReturns[i,j]=((stockPrice[i+1, j]-stockPrice[i,j])/stockPrice[i,j])*100

#     return stockReturns

def getCompoundReturnLog(dataPrices):   
    returns = np.log(dataPrices/dataPrices.shift(1)).dropna(how="any")

    return np.asarray(returns)
```

```python
# pegando as labels das stocks
stockLabels = dfStock.columns[1:stockCols + 1].tolist()

# compute stock returns
stockReturnsArr = getCompoundReturnLog(stockData)*100

# np.set_printoptions(precision = 3, suppress = True)

meanReturns = np.mean(stockReturnsArr, axis = 0)
print('Mean returns:\n', meanReturns)

covReturns = np.cov(stockReturnsArr, rowvar = False)
print('Variance-Covariance Matrix of returns:\n', covReturns)
print(len(covReturns))
```

```python
# consideramos que as stocks com Beta > 1 são as mais voláteis, temos a seguinte separação entre elas

# classificando stocks em alto e baixa volatilidade

from utils import getVolatilityArr, SLSQPSSolver

new_beta = [0.782, 0.704, 1.753, 1.257, 0.733, 1.166, 1.173, 0.558, 1.016, 1.150, 0.729, 0.788, 0.988, 1.712, 0.676, 0.960, 0.637, 1.939, 1.643, 1.280, 1.183, 1.114, 1.481, 0.546, 0.822, 0.959, 0.760, 1.642, 0.749, 0.964, 0.943, 1.651, 0.985,\
            0.912, 0.821, 0.942, 0.936, 1.113, 0.367, 1.486, 0.416, 1.226, 1.307, 1.180, 1.582, 0.168, 0.543, 1.393, 1.252, 1.055, 1.197, 1.229, 1.219, 0.948, 0.678, 1.007, 0.794, 0.946, 1.418, 1.419, 0.802, 0.720, 1.444]
# volatArr eh a "classificacao" das stocks entre volateis ou nao
[volatArr, boundariesArr] = getVolatilityArr(new_beta,[0.1, 0.01])

portiSize = 63 # numero de stocks possiveis

# init
optimalPorti = []
minRiskPoint = []
expPortifolioReturnPoint = []

for points in range(0,100):
    riskAversParam = points/100.0
    result = SLSQPSSolver(meanReturns, covReturns, riskAversParam, portiSize, volatArr, boundariesArr)
    optimalPorti.append(result.x)
    
xOptimumArr = np.array(optimalPorti)
minRiskPoint = np.diagonal(np.matmul((np.matmul(xOptimumArr,covReturns)), np.transpose(xOptimumArr)))

riskPoint = np.sqrt(minRiskPoint*251)
expPortfolioReturnPoint= np.matmul(xOptimumArr, meanReturns)
retPoint = 251 * np.array(expPortfolioReturnPoint)

np.set_printoptions(precision=3, suppress=True)

print("Optimal weights of the efficient set portfolios\n:", xOptimumArr)
print("\nAnnualized Risk and Return of the efficient set portfolios:\n",\
      np.c_[riskPoint, retPoint])
```
